```python
import numpy as np
```

```python
def get_confusion_matrix(y_hat, y_true, num_classes):
  confusion_matrix = np.zeros((num_classes, num_classes))
  
  for i in range(len(y_hat)):
    confusion_matrix[y_true[i]][y_hat[i]] += 1
    
  return confusion_matrix
```

```python
def precision(confusion_matrix, idx):
  return confusion_matrix[idx][idx] / confusion_matrix[:, idx].sum()
```

```python
def recall(confusion_matrix, idx):
  return confusion_matrix[idx][idx] / confusion_matrix[idx, :].sum()
```

```python
def f1_score(precision, recall):
  return 2 * ((precision * recall) / (precision + recall))
```

```python
def accuracy(confusion_matrix):
  return confusion_matrix.diagonal().sum() / confusion_matrix.sum()
```

```python
def get_classification_report(y_hat, y_true, num_classes):
  
  confusion_matrix = get_confusion_matrix(y_hat, y_true, num_classes)
  
  acc = accuracy(confusion_matrix)
  
  if(num_classes > 2):
    r = np.zeros((num_classes))
    p = np.zeros((num_classes))
    f1 = np.zeros((num_classes))
    
    for i in range(num_classes):
      r[i] = recall(confusion_matrix, i)
      p[i] = precision(confusion_matrix, i)
      f1[i] = f1_score(p[i], r[i])
    
    r = r.mean()
    p = p.mean()
    f1 = f1.mean()
  else:
    r = recall(confusion_matrix, 1)
    p = precision(confusion_matrix, 1)
    f1 = f1_score(p, r)
  
  return acc, r, p, f1
```

```python
y_pred = np.array([1, 1, 0, 0, 1, 2])
y_true = np.array([0, 0, 0, 0, 1, 1])

get_classification_report(y_pred, y_true, 3)
```
